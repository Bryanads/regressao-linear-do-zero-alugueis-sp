```python
import pandas as pd
df = pd.read_csv('../data/02_data.csv')
df.head()
df.describe()
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt

colunas = ['district_te', 'type_te']
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
axes = axes.flatten()
for i, var in enumerate(colunas):
    sns.histplot(df[var], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribuição de {var}')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
axes = axes.flatten()
for i, var in enumerate(colunas):
    sns.boxplot(data=df, x=var, ax=axes[i])
    axes[i].set_title(f'Boxplot de {var}')
plt.tight_layout()
plt.show()
```

```python
colunas = ['rent', 'bedrooms', 'garage', 'area', 'district_te', 'type_te']
dados_numericos = df[colunas]
sns.heatmap(dados_numericos.corr(), annot=True, cmap='coolwarm')
plt.title('Correlação entre variáveis')
plt.show()
```
